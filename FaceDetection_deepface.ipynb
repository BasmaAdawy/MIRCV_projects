```python
from deepface import DeepFace
import cv2
import urllib.request
import numpy as np

# URL of the image from the internet
image_url = "https://res.cloudinary.com/demo/image/upload/w_0.5,ar_1,c_thumb,g_faces,z_0.7/r_max/docs/young-couple.jpg"

# Download the image
req = urllib.request.urlopen(image_url)
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1)  # 'Load it as it is'

# Analyze the image
results = DeepFace.analyze(img, actions=['emotion', 'age', 'gender', 'race'])

# Print results
print("Dominant Emotion:", results[0]['dominant_emotion'])
print("Age:", results[0]['age'])
print("Gender:", results[0]['gender'])
print("Dominant Race:", results[0]['dominant_race'])
```

```python
from deepface import DeepFace
from ultralytics import YOLO
import cv2

# Step 1: Load the image once (outside the loop)
image_path = "/content/young-couple.jpg"
image = cv2.imread(image_path)

if image is None:
    raise FileNotFoundError(f"Image not found at {image_path}")

# Step 2: Detect faces with YOLO
model = YOLO("yolov8n.pt")  # or a custom face-detection model
results = model(image)

# Step 3: Crop faces and analyze age/gender
for box in results[0].boxes.xyxy:
    x1, y1, x2, y2 = map(int, box)
    face_img = image[y1:y2, x1:x2]  # Crop from the already-loaded image

    # Check if the face region is valid
    if face_img.size == 0:
        print("Warning: Empty face region detected!")
        continue

    try:
        demography = DeepFace.analyze(face_img, actions=["age", "gender"])
        print(f"Age: {demography[0]['age']}, Gender: {demography[0]['gender']}")
    except Exception as e:
        print(f"Error analyzing face: {e}")
```

```python
from deepface import DeepFace
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Required for Colab

# Step 1: Load the image
image_path = "/content/young-couple.jpg"
image = cv2.imread(image_path)

if image is None:
    raise FileNotFoundError(f"Image not found at {image_path}")

# Step 2: Detect faces with YOLO
model = YOLO("yolov8n.pt")  # or a custom face-detection model
results = model(image)

# Step 3: Process each detected face
for box in results[0].boxes.xyxy:
    x1, y1, x2, y2 = map(int, box)
    face_img = image[y1:y2, x1:x2]  # Crop face

    if face_img.size == 0:
        continue  # Skip empty regions

    # Analyze age & gender
    try:
        demography = DeepFace.analyze(face_img, actions=["age", "gender"], enforce_detection=False)
        age = demography[0]['age']
        gender = demography[0]['gender']

        # Draw bounding box and labels
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green box
        label = f"{gender}, {age}"
        cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

        print(f"Detected: {label}")
    except Exception as e:
        print(f"Error analyzing face: {e}")

# Step 4: Display the image with annotations (Colab-friendly)
cv2_imshow(image)  # Replaces cv2.imshow()
```
